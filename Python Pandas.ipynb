```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'First Name': ['Shiv', 'Aadi', 'Satish', 'Rana'],
                   'Last Name': ['Patel', 'Patel', 'Panchal', 'Aniket'],
                   'Type Name': ['Full Time Emp.', 'Intern', 'Full Time Emp.', 'Part-Time Emp.'],
                   'YoE': [2, 3, 4, 6],
                   'Department': ['Administration', 'Technical', 'Management', 'Swepper Staff'],
                   'Salary': [20000, 40000, 10000, 1000]})
df
```

```python
df.pivot(index="Department", columns="YoE", values="Salary")
```

```python
df.pivot(index="First Name", columns="Type Name", values="YoE")
```

```python
output = pd.pivot_table(data=df,
                       index=['Type Name'],
                       columns=['Department'],
                       # value='Salary',
                       aggfunc='mean') #aggfunc is the aggregate function for mean
output
```

```python
output = pd.pivot_table(data=df,
                       index=['Type Name'],
                       columns=['Department'],
                       values='Salary',
                       aggfunc='mean') #aggfunc is the aggregate function for mean
output
```

```python
# Calculate row and column totals (margins)

output = pd.pivot_table(data=df, index=['Type Name'],
                       #columns=['Department'],
                       values='Salary',
                       aggfunc=['mean', 'sum', 'count', 'max', 'min'],
                       margins=True,
                       margins_name='Total'
                       ) #aggfunc is the aggregate function for mean
output
```

```python
output = pd.pivot_table(data=df, index=['Type Name'],
                       #columns=['Department'],
                       values=['YoE', 'Salary'],
                       aggfunc=['mean', 'sum', 'count', 'max', 'min'],
                       margins=True,
                       margins_name='Total'
                       ) #aggfunc is the aggregate function for mean
output
```

```python
shiv = pd.pivot_table(data=df, index=['Type Name'],
                       columns=['Department'],
                       values=['YoE', 'Salary'],
                       aggfunc={'Salary': np.sum, 'YoE': np.mean})
shiv
```

```python
shiv = pd.pivot_table(data=df, index=['Type Name'],
                       columns=['Department'],
                       values=['YoE', 'Salary'],
                       aggfunc={'Salary': np.sum, 'YoE': np.mean},
                       fill_value='NA')
shiv
```

```python
output = pd.pivot_table(data=df, index=['Type Name', 'First Name', 'Last Name'],
                       #columns=['Department'],
                       values=['Salary'],
                       aggfunc=['mean', 'sum', 'count'],
                       margins=True,
                       margins_name='Grand Total'
                       ) #aggfunc is the aggregate function for mean
output
```

```python
import pandas as pd

data = {'Unemp_rate': [6.1,5.2,5.7,3.8,5.6,4.6,6.6],
        'stock_index_price': [1500,2000,1520,1560,1800,1750,1680]}

df = pd.DataFrame(data,columns=['Unemp_rate', 'stock_index_price'])
print(df)
```

```python
import matplotlib.pyplot as plt

df.plot(x = 'Unemp_rate', y = 'stock_index_price', kind = 'scatter')
plt.show();
```

```python
df.plot(x = 'Unemp_rate', y = 'stock_index_price', kind = 'line', marker ='+', c = 'r')
plt.show();
```

```python
data = {'country': ['USA', 'Canada', 'Germany', 'UK', 'India'],
       'GDP_Per_Capita': [45000, 42000, 52000, 49000, 60000]}

df = pd.DataFrame(data,columns=['country', 'GDP_Per_Capita'])
print(df);
# plotting the points
plt.bar()

# naming the x axis
plt.xlabel('country')

# naming the y axis
plt.ylabel('GDP_Per_Capita')

# Giving the title Name
plt.suptitle("TITLE")
```
