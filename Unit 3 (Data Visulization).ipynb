```python
#draw the plot as fig

fig, ax = plt.subplots()   #ax is the variable (we can )

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

#text annotation to the plot where it indicate maximum value of the curv
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

#text annotation to the plot where it indicate minimum value of the curv
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3, angleA=0,angleB=-90"))
```

# Adding Annotate & Legend

```python
import matplotlib.pyplot as plt
import numpy as np
#You can also write
#from matplotlib import pyplot as plt

x = np.arange(0, 10)
y = x^2
z = x^3

# plotting the points


plt.plot(x,y)
plt.plot(y,z)
plt.plot(x,z)
plt.legend(["Shiv", "Aaditya", "Satish"], loc=0)

plt.title("RESULT")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.yticks(rotation=60)


#Annotation

plt.annotate(xy=[2,1],text="First Entry", rotation=60) # Indicate the position or co-ordinate first
plt.annotate(xy=[4,6],text="Second Entry", rotation=30)
plt.annotate(xy=[8,10],text="Third Entry", rotation=20)

plt.annotate(xy=[0,1],text="x")
plt.annotate(xy=[4,5],text="y")
plt.annotate(xy=[10,11],text="z")

plt.annotate(xy=[0,3],text="C")
plt.annotate(xy=[4,7],text="S")
plt.annotate(xy=[8,11],text="P")

plt.show()
```

# Pie Plot

pie chart is a circular statastical plot that can display only one series of data.

The area of the chart is the totle % of the given data.

```python
# importing the required modules
import matplotlib.pyplot as plt
import numpy as np

#Creating dataset

Branch = ["Computer", "Civil", "Electrical", "Mechanical", "Automobile", "IT"]

data = [50,34,40,30,26,12]
myexplode = [0.2, 0, 0, 0.1, 0, 0]
mycolors = ["orange", "hotpink", "blue", "green", "yellow", "red"]

#Creat chart
fig = plt.figure(figsize=(10,7))
plt.pie(data, labels=Branch, startangle = 90, explode = myexplode, autopct='%1.1f%%', shadow = True, colors = mycolors)
plt.legend()

plt.show()
```

# 3D ploting

```python
# importing the required modules
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits import mplot3d #3d toolkit lib

```

```python
fig = plt.figure()

#syntax for 3D projection

ax = plt.axes(projection = '3d')
             

# defining all 3 axes
z = np.linspace(0, 10, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

# plotting
ax.plot3D(x, y, z, 'green')
ax.set_title('3D')
plt.show()
```
